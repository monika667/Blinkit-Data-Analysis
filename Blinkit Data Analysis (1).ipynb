### **BUSINESS REQUIREMENTS**

### **KPI's REQUIREMENTS**

```python
# Convert Sales and Rating to numeric
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Total Sales
total_sales = df['Sales'].sum()

# Average Sales
avg_sales = df['Sales'].mean()

# No. of Items Sold (assuming each row is a sale)
no_of_items_sold = df['Sales'].count()

# Average Ratings
avg_rating = df['Rating'].mean()

# Display Results
print(f"Total Sales: {total_sales:.0f}")
print(f"Average Sales: {avg_sales:.1f}")
print(f"No Of Items Sold: {no_of_items_sold:.0f}")
print(f"Average Rating: {avg_rating:.1f}")
```

### **CHARTS REQUIREMENTS**

### **1. Total Sales by Fat Content**

**Q. What percentage of sales does "Low Fat" content represent, and what does "Regular" represent?** 

*  "Low Fat" represents 65% of sales, while "Regular" represents 35%.

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat,
        labels=sales_by_fat.index,
        autopct='%.0f%%',
        startangle=90,
        wedgeprops={'width': 0.4})  # Creates the donut hole

plt.title('Sales by Fat Content')
plt.axis('equal')  # Keeps the circle circular
plt.show()
```

### Insights:
**Percentage of Total Sales by Fat Content(Donut Chart)**

* Low Fat items generate 65% of total sales, indicating strong customer preference for healthier options.

* Regular Fat items contribute 35%, showing relatively lower demand in comparison.

### **2. Total Outlet Location Type Count by Item Type**

**Q. Which item type had the highest Outlet Location Type Count, and what was that amount?**
 *  Fruits and Vegetables had the highest total sales at 1,232.

```python
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip()  # remove leading/trailing spaces
sales_by_type_total = df.groupby('Item Type')['Outlet Location Type'].count().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type_total.index, sales_by_type_total.values)
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Count of Outlet Location Type')
plt.title('Total Outlet Location Type Count by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

### Insights:
**Total Sales by Item Type(Bar Chart)**

* Fruits & Vegetables and Snack Foods are the top-selling item types.

* Seafood and Breakfast items have the lowest sales, indicating lower demand or limited availability.

### **3. Fat Content by Outlet Tier for Total Sales**

**Q. In which outlet tier do "Low Fat" items have the highest total sales, and approximately how much are those sales?** 

* "Low Fat" items have the highest total sales in Tier 3, with approximately 280,000 in sales.

### **4. Stacked Sales by Outlet Size and Item Fat Content**

### **7. Average Item Weight by Outlet Type**

**Q. Which outlet type has the lowest average item weight, and which has the highest?**

* Supermarket Type1 has the lowest average item weight, while Grocery Store has the highest.

```python
import matplotlib.pyplot as plt

# Drop rows with missing or non-numeric 'Item Weight' values
df['Item Weight'] = pd.to_numeric(df['Item Weight'], errors='coerce')  # Convert to numeric (non-numeric becomes NaN)
df_clean = df.dropna(subset=['Item Weight'])  # Drop rows with NaN in 'Item Weight'

# Group and calculate mean
avg_weight_by_outlet = df_clean.groupby('Outlet Type')['Item Weight'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_weight_by_outlet.index, avg_weight_by_outlet.values, marker='o', linestyle='-', color='red')
plt.title('Average Item Weight by Outlet Type', fontsize=14)
plt.xlabel('Outlet Type', fontsize=12)
plt.ylabel('Average Item Weight', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Insights:
**Average Item Weight by Outlet Type(Line Chart)**

* Grocery Stores have the highest average item weight, while Supermarket Type1 has the lowest (12.93).

* The variation in average item weight is minimal but suggests slight differences in product types across outlet formats.

### **8. Total Sales by Outlet Location Type**

**Q. Which outlet location type generates the highest total sales, and which generates the lowest?** 

* Tier 3 generates the highest total sales, and Tier 1 generates the lowest.

### Insights:
**Total Sales by Outlet Location Type( Barplot Chart)**

* Tier 3 outlets generate the highest total sales, surpassing both Tier 1 and Tier 2 locations.

### **9. Outlet Size Distribution by Item Fat Content**

**Q. Which outlet size category has a larger proportion within "Low Fat" items compared to "Regular" items?** 

* The "Small" outlet size category appears to have a larger proportion within "Low Fat" items compared to "Regular" items. 

```python
import pandas as pd
import matplotlib.pyplot as plt

outlet_data = pd.crosstab(df['Item Fat Content'], df['Outlet Size'])
outlet_data_percent = outlet_data.div(outlet_data.sum(axis=1), axis=0) * 100

outlet_data_percent.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Pastel1')
plt.title('100% Stacked Column Chart: Outlet Size Distribution by Item Fat Content', fontsize=14)
plt.xlabel('Item Fat Content', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.legend(title='Outlet Size')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Insights:
**Outlet Size Distribution by Item Fat Content(100% Stacked Column Chart)**
* Medium-sized outlets dominate both Low Fat and Regular item categories in distribution.

* High-sized outlets are more common with Regular items than with Low Fat items
